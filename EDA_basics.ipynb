```python
import pandas as pd 
link ='https://docs.google.com/spreadsheets/d/e/2PACX-1vSErLAEGI-BOimrZbvL461C2BoUXoOY4Z_Kye-pdfSQ9sHfhaCVS4dQiCfRnmRRDa0uAVVtf9slUtGL/pub?gid=1842437627&single=true&output=csv'
df = pd.read_csv(link)
```

```python
df.iloc[:, 4:] = df.iloc[:, 4:].apply(pd.to_numeric, errors='coerce')

df = df[df['Country Name'] != 'World']

# Stack the year columns and find the overall min and max values
stacked = df.set_index(['Country Name', 'Country Code']).iloc[:, 2:].stack()
print(stacked)
overall_min_value = stacked.min()
overall_max_value = stacked.max()

# Find the countries with overall min and max values
overall_min_country = stacked.idxmin()[0]
overall_max_country = stacked.idxmax()[0]

# Find min and max values for 2023
min_2023_value = df['2023'].min()
max_2023_value = df['2023'].max()

# Find the countries with min and max values in 2023
min_2023_country = df[df['2023'] == min_2023_value]['Country Name'].values[0]
max_2023_country = df[df['2023'] == max_2023_value]['Country Name'].values[0]


# Output results
print(f"Overall Min Value: {overall_min_value} (Country: {overall_min_country})")
print(f"Overall Max Value: {overall_max_value} (Country: {overall_max_country})")
print(f"2023 Min Value: {min_2023_value} (Country: {min_2023_country})")
print(f"2023 Max Value: {max_2023_value} (Country: {max_2023_country})")
```
