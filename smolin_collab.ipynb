## Option 2: via GoogleCredentials


```python
# Install the required libraries
#!pip install gspread google-auth

# Import libraries
import gspread
from gspread_dataframe import get_as_dataframe
from google.colab import auth
from oauth2client.client import GoogleCredentials
from google.auth import default
import pandas as pd

# Authenticate and create a client
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# take the excel spreadsheet with mutiple worksheet tabs that I shared, upload/ add it to your drive and open it in google sheets.
# then paste the link to that sheet here:

spreadsheet_url = 'https://docs.google.com/spreadsheets/d/13IZxy4XwbQ7dXxGmj0ODostAD_mJf9S3rhHXrdYtO0I/edit?gid=0#gid=0'

# Open the spreadsheet
spreadsheet = gc.open_by_url(spreadsheet_url)
```

```python
# Get all the worksheet objects (tabs) in the spreadsheet
worksheets = spreadsheet.worksheets()

# Create a dictionary of DataFrames for each sheet
dfs = {}
for sheet in worksheets:
    sheet_name = sheet.title
    df = get_as_dataframe(sheet)
    dfs[sheet_name] = df

# Print the names of the sheets and their respective DataFrame head
for sheet_name, df in dfs.items():
    print(f"name of sheet: {sheet_name}")
    #print(df.head())  # Show first few rows of each DataFrame - NB it contains tons of NaNs
```

```python
df1 = dfs['name1']
print('initial shape:', df1.shape)
#print(df1)
df1.columns = df1.columns.astype(str)
df1 = df1.loc[:,~df1.columns.str.match("Unnamed")]
print('after removing unndamed columns: ', df1.shape)
df1 = df1.dropna()
print('after removing rows with NaNs: ', df1.shape)
df1
```

```python
df2 = dfs['name2']
print('initial shape:', df2.shape)
df2 = df2.loc[:,~df2.columns.str.match("Unnamed")]
print('after removing unndamed columns: ', df2.shape)
df2 = df2.dropna()
print('after removing rows with NaNs: ', df2.shape)
df2
```
